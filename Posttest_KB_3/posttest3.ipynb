<b> Diagram Batang </b>

```python
x = df["genre"]
y = df["budget"]

plt.title("Perbandingan Budget Movies Berdasarkan Genre")
plt.xlabel("BUDGET")
plt.ylabel("GENRE")

plt.barh(x, y, color="violet")
plt.show()
```

<b> Diagram Garis </b>

```python
# ambil semua record yg di kolom country ada United States
data = df[(df["country"].isin(["United States"])) & (df["year"] > 2015)]

# dari data, ambil kolom year dan budget
x = data["year"]
y = data["budget"]

plt.title("Perkembangan Budget Movies United States dari Tahun ke Tahun")
plt.xlabel("YEAR")
plt.ylabel("BUDGET")

plt.plot(x, y)
plt.show()
```
