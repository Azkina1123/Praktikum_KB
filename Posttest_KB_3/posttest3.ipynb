<b> Diagram Garis </b> <br>
Pada diagram ini, akan divisualisasikan perkembangan Budget yang digunakan untuk produksi film tiap tahunnya sejak 1980-2020

Cari rata-rata Budget yang digunakan per tahunnya.

```python
mean_budget = df[["year", "budget"]].groupby("year").mean()

# tampilkan beberapa data apakah sudah benar
mean_budget.head()
```

Visualisasikan data rata-rata budget per tahun dari tahun 1980-2020.

```python
mean_budget.plot(
    kind="line",
    xlabel="Tahun",
    ylabel="Rata-Rata Budget",
    title="Rata-Rata Budget untuk Film per Tahun",
    figsize=(10, 5),
    color="black"
)

plt.show()
```

Dari diagram di atas, dapat disimpulkan bahwa Budget yang dikeluarkan untuk produksi film dari tahun ke tahun cenderung meningkat.

<b> Scatter Plot </b> <br>
Pada diagram ini, akan divisualisasikan hubungan antara Budget yang dikeluarkan untuk film dengan Score yang didapatkan pada film tersebut.

```python
plt.title("Perbandingan Besarnya Budget dengan Tingginya Score pada Film")
plt.xlabel("Budget")
plt.ylabel("Score")

sns.scatterplot(
    data=df, 
    x="budget",
    y="score",
    color="green"
)
plt.show()
```

Pada scatter plot di atas, dapat disimpulkan bahwa besarnya <b>budget</b> yang dikeluarkan belum tentu akan dapat memuaskan penonton dan mendapatkan <b>score</b> yang tinggi. Hal ini terbukti terdapat beberapa film yang mempunyai <b>budget</b> yang lebih kecil, tetapi memiliki <b>score</b> yang lebih tinggi daripada film yang memiliki <b>budget</b> terbesar.

<b> Diagram Batang </b> <br>
Pada diagram ini, akan divisualisaikan tingkat kepuasan penonton terhadap suatu film berdasarkan <b>genre</b> dengan mengambil indikator <b>score.</b>

Ambil atribut yang diperlukan dari dataframe utama.

```python
# ambil atribut genre, score, dan budget
data = df[["genre", "score"]]

data
```

Cari rata-rata <b>score</b> dari pada tiap <b>genre</b>.

```python
data = data[["genre", "score"]].groupby("genre").mean()

data
```

Tampilkan diagram batang.

```python
data.plot(
    kind="bar",
    xlabel="Genre",
    ylabel="Score",
    figsize=(10, 5),
    color=["grey"]
)

plt.show()
```

Pada diagram batang di atas dapat dilihat bahwa <b>genre</b> History dan Musical yang mendapatkan <b>score</b> tinggi.

<b>Grafik Pie</b> <br>
Pada grafik ini, akan ditampilkan <b>genre</b> apa saja yang diproduksi di <b>country</b> United States pada tahun 2020.

Ambil atribut dan record yang dibutuhkan untuk divisualisasikan.

```python
data = df[(df["country"].isin(["United States"])) & (df["year"] == 2020)][["year", "genre"]]

data
```

Tampilkan grafik pie.

```python
data_pie = data["genre"].value_counts().rename_axis("genre").reset_index(name="id_count")

plt.figure(figsize=(10, 10))
plt.pie(
    data_pie["id_count"], 
    labels=data_pie["genre"],
    startangle=90,
    autopct="%.1f%%"
)
plt.title("Genre yang Memproduksi Film pada tahun 2020")
plt.show()
```

Pada grafik pie di atas, didapatkan informasi bahwa negara United States lebih banyak memproduksi film ber<b>genre</b> Action pada tahun 2020.
